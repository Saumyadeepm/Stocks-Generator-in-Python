*• Create a Class for storing Stock Price. A stock price will have the time of observation, open value, high value, low value, and current value (we are not considering the close value for this problem).
<br>*
*• The observation time will be generated based on the actual time that the object is created (system's time).
<br>
• Other values could be created through Constructors, after giving a prompt to Users.
<br>
• Create Getter and Setter methods for any of the parameters.
<br>
• Define a toString method which returns the current values of an object, as a dictionary.*

```python
#Creating Class

class StockPrice:
    companyName = "Samsung"
    def __init__(self, stockname, timeObv, openv, highv, lowv, currentv, sno):
        self.stockname = stockname
        self.timeObv = timeObv
        self.openv = openv
        self.highv = highv
        self.lowv = lowv
        self.currentv = currentv
        self.sno = sno
    
        
    
    def display(self):
        print("*"*25)
        print("Time Of Observation : " + str(self.timeObv))
        print("Opening Value       : ₹" + str(self.openv))
        print("High Value          : ₹" + str(self.highv))
        print("Low Value           : ₹" + str(self.lowv))
        print("Current Value       : ₹" + str(self.currentv))
        print("-"*25)
        print("Company Name        : " + str(self.stockname))
        print("Stock No.           : " + self.sno)
        
        
    def __setattribute__(self, stockname, timeObv, openv, highv, lowv, currentv):
        self.stockname = stockname
        self.timeObv   = timeObv   #instance Attributes
        self.openv     = openv
        self.highv     = highv
        self.lowv      = lowv
        self.currentv  = currentv 
        
    
    def setToOpenv(self,new_Openv):     #Method for "To Change Opening Value"
            self.openv = new_Openv

    def setToLowv(self,new_Lowv):     #Method for "To" Opening Value
            self.lowv = new_Lowv
    
    def getOpenv(self):
        return self.openv
    
    def getHighv(self):
        return self.highv
    
    def getLowv(self):
        return self.lowv
    
    def getCurrentv(self):
        return self.currentv
```

```python
##User Defined Functions For Generating Stock Names & Prices##

import string
import random



def stock_name_generator(num):
    names = []

    for n in range(num):
        x = ''.join(random.choices(string.ascii_uppercase, k=4))
        names.append(x)

    return names    
print(stock_name_generator(1))

###################
random.seed(0)


def stock_price_generator(n=1):
    return [(random.random()) for _ in range(n)]


stocks = stock_price_generator()

print(stocks)
stocks[0]


from datetime import datetime

# datetime object containing current date and time
def timenow():
    now = datetime.now()
    n= str(now.strftime("%d/%m/%Y %H:%M:%S"))    
    return n
```
